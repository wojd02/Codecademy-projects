```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Converting the datasets

```python
species = pd.read_csv('species_info.csv')
observations = pd.read_csv('observations.csv')
```

### Species

The `species_info.csv` contains information on the different species in the National Parks. The columns in the data set include:
- **category** - The category of taxonomy for each species
- **scientific_name** - The scientific name of each species
- **common_names** - The common names of each species
- **conservation_status** - The species conservation status

```python
display(species.head())
```

### Observations

The `Observations.csv` contains information from recorded sightings of different species throughout the national parks in the past 7 days. The columns included are:

- **scientific_name** - The scientific name of each species
- **park_name** - The name of the national park
- **observations** - The number of observations in the past 7 days

```python
display(observations.head())
```

```python
print(species.dtypes)
print()
print(observations.dtypes)
```

## Detailing each species and its situation in more detail

I filtered the tables to get the exact number of species in each status and then checked if the graph was accurate.

The `NaN` values means that these species are not on list of concerns.

```python
display(species.conservation_status.value_counts())
print()
display(species.conservation_status.isna().sum())
```

I decided to change the NaN values to 'OK' to simplify the code in the future.

```python
#fill the NaN values:
species.fillna('OK', inplace=True)
#checking values again:
display(species.conservation_status.value_counts())
```

I created an overview chart to show the **distribution of groups** in their respective conservation statuses.

```python
category_status = species[species.conservation_status != 'OK'].groupby(['conservation_status', 'category'])['scientific_name'].count().unstack()
display(category_status)
```

```python
ax = category_status.plot(kind='bar', stacked=True)
plt.xticks(rotation=0)
plt.ylabel('Number of species')
plt.show()
plt.clf()
```

```python
sep_species = species.category[species.conservation_status == 'Endangered'].value_counts()
print(sep_species)
```

I generated separate graphs so I could better understand the relations between `category` X `conservation status`.

```python
conservation_status_values = ['Species of Concern', 'Endangered', 'Threatened', 'In Recovery']
for status in conservation_status_values:
    species.category[species.conservation_status == status].value_counts().plot(kind = 'bar')
    plt.xticks(rotation=45)
    plt.title(f'Number of {status} by category')
    plt.show()
    plt.clf()
```

## Protected X Not protected species

Table showing the number of `protected` and `not_protected` species and their percentages.

```python
species['protected'] = species.conservation_status != 'OK'
pct_status = species.groupby(['category', 'protected']).scientific_name.nunique().reset_index().pivot(columns='protected', index='category', values='scientific_name').reset_index()
pct_status.columns = ['category', 'not_protected', 'protected']

pct_status['percent_not_protected'] = pct_status.not_protected / (pct_status.not_protected + pct_status.protected) * 100
display(pct_status)
```

```python
from scipy.stats import chi2_contingency

#p value of Mammal X Reptile
crosstable_1 = [[30, 146],[5, 73]]
chi, p_val, dof, expected_table = chi2_contingency(crosstable_1)
print(f'The p value of Mammal and Reptile is {p_val:.2f}')

#p value of Mammal X Birds
crosstable_2 = [[30, 146],[75, 413]]
chi2, p_val2, dof, expected_table2 = chi2_contingency(crosstable_2)
print(f'The p value of Mammal and Birds is {p_val2:.2f}')
```

A p-value below 0.5 indicates that there is a relationship between the variables, while values above 0.5 indicate that the relationship is not as significant. The p-value for mammals relative to reptiles indicates that mammals require greater protection compared to reptiles.

## Joining the datasets

To get a better idea of the number of observations of these groups in each park, I thought it would be a good option to combine using the column `scientific_name` of the two datasets and some filters to arrive at the result I was expecting.

```python
species_obs = pd.merge(species, observations, on='scientific_name', how='inner')
species_obs.head()
```

## Pivoting the dataset:

I removed the line for `Vascular Plants` because the discrepant numbers compared to the other groups were distorting the graphs and it was impossible to see the exact numbers for the remaining groups.

I also wanted to see the total number of sightings of each species in different parks. So I pivoted the merging of the two datasets to focus only on the data that interested me at the moment.

```python
view_per_park = species_obs.pivot_table(columns = 'park_name',index = 'category', values = 'observations', aggfunc='sum')
view_per_park = view_per_park.drop('Vascular Plant')
display(view_per_park)
```

## Groups observed in the parks

The largest number of observations in each park, all occupied by the vascular plant category.

```python
most_view = species_obs.groupby('park_name')['observations'].max()
most_view = most_view.tolist()
print(most_view)
parks_names = ['Bryce National Park', 'Great Smoky Mountains National Park', 'Yellowstone National Park', 'Yosemite National Park']
```

```python
for x in range(4):
    display(species_obs[(species_obs.park_name == parks_names[x]) & (species_obs.observations == most_view[x])])
```

```python
for park in parks_names:
    view_per_park[park].plot(kind = 'pie', autopct='%.2f%%')
    plt.title(f'Most species spotted at {park}')
    plt.show()
    plt.clf()
```
