# Importing required libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### importing the dataset

```python
data = pd.read_csv('all_data.csv')
data.head(10)
```

### Columns adjust

```python
data = data.rename(columns={'Country': 'country', 'Year': 'year', 'Life': 'life', 'Life expectancy at birth (years)': 'life_expectancy_at_birth_years', 'GDP': 'GDP'})
data.head()
```

### Exploring the data

```python
print(data.country.unique())
print()
years = data.year.unique()
countries = data.country.unique()
print(years)
print()
print(data.dtypes)
```

```python
#plot life expectancy X years
for country in countries:
    plt.plot(years, data.life_expectancy_at_birth_years[data.country == country])

plt.legend(data.country.unique(), bbox_to_anchor=(1.05, 1), loc='upper left') #bbox_to_anchor(xloc, yloc)
plt.xlabel('years')
plt.ylabel('Life expectancy')
plt.title('Life expectancy over time')

plt.show()
plt.clf()
```

```python
#plot GDP X years
for country in countries:
    plt.plot(years, data.GDP[data.country == country])

plt.legend(data.country.unique(), bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel('GDP')
plt.title('GDP over the time')

plt.show()
plt.clf()
```

```python
#correlation between GDP and Life expectancy
grade_GDP_life_exp = sns.FacetGrid(data, col="country", col_wrap=3,
                      hue = "country", sharey = False, sharex = False)

grade_GDP_life_exp = (grade_GDP_life_exp.map(sns.lineplot,"life_expectancy_at_birth_years", "GDP")
         .add_legend()
         .set_axis_labels("Life expectancy at birth (years)","GDP"))

grade_GDP_life_exp;
```

```python
for country in countries:
    # Get data for this specific country
    country_data = data[data.country == country]
    # Use the mean life expectancy for each country for the bar chart
    plt.bar(country, country_data.life_expectancy_at_birth_years.mean())

plt.xticks(rotation=30)  # Rotate country names for better readability
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy by Country')

plt.tight_layout()  # Adjust layout to make room for rotated labels

plt.show()
plt.clf()
```
